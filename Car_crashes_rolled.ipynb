```python
# preview the data
pd.set_option('display.max_colwidth', 100)
crashes
```

```python
# searches for the word 'roll' in a string
def car_rolls(cell):
    index = cell.find('roll')
    if index == -1:
        return False
    return True
```

```python
# create new column that indicates whether a car rolled in the crash
crashes['ROLLED'] = crashes['SUMMARY'].apply(car_rolls)
```

```python
# preview the edited dataframe
crashes
```

```python
# create a new data frame that only consists of the crashes where a car rolled
rolled_crashes = crashes[crashes['ROLLED'] == True]
```

```python
# preview it
rolled_crashes
```

```python
# drop the 'ROLLED' column as it is now redundant
rolled_crashes.drop('ROLLED',axis=1)
```

```python
# get the percentage of how many crashes involved a car rolling
rolled_crashes.shape[0]/crashes.shape[0]
```
